## Algoritmos de Aprendizaje de Maquina

Dentro de los aspectos de computación científica los algoritmos de aprendizaje de maquina son usadas para poder realizar procesos que incurran en aprendizaje, inteligencia y predicción con el fin de poder dar soluciones a problemas reales y complejos que necesitan de estos elementos, dentro de los algoritmos de aprendizaje de maquina existe una clasificación entre supervisados y no supervisados, vamos a evidenciar un modelo para el aprendizaje supervisado un ejemplo de **regresión lineal**

### Aprendizaje Supervisado: Regresión Lineal

El modelo de regresión se usa para predecir o aproximar un valor real u observado del número **y** que está asociado con **x**

```julia
# Parametros de la regresión lineal
beta = [148.73, -18.85]; v = 54.40;
y_hat(x) = x'*beta + v;
```

```julia
# Evaluar la predicción del modelo de regresión
x = [0.846, 1]; y = 115;
y_hat(x), y
```

```julia
x = [1.324,2]; y = 234.50;
y_hat(x), y
```

Podemos notar que nuestra primera predicción no es muy precisa, pero la segunda tiene una mejora significativa

### Predicción usando datos de ventas de casas

Un diagrama de dispersión de los precios predichos y reales de la vivienda puede ser generado de la siguiente manera. Utilizamos la función VMLS **house_sales_data** para obtener
el precio de los vectores, área, camas

```julia
using VMLS
```

```julia
# Cargando en la variabla D la información correspondiente a los datos de ventas de casas
D = house_sales_data();
```

```julia
# Obteniendo los precios de las casas en la variable price
price = D["price"];
```

```julia
# Obteniendo la cantidad de camas de las casas en la variable beds
beds = D["beds"];
```

```julia
# Obteninedo el area de cada una de las casas en la variable area
area = D["area"];
```

```julia
v = 54.4017;
```

```julia
beta = [ 148.7251, -18.8534 ];
```

```julia
predicted = v .+ beta[1] * area + beta[2] * beds;
```

```julia
using Plots
```

```julia
scatter(price, predicted, lims = (0,800));
```

```julia
plot!([0, 800], [0, 800], linestyle = :dash);
```

```julia
# make axes equal and add labels
plot!(xlims = (0,800), ylims = (0,800), size = (500,500))
plot!(xlabel = "Precio Actual", ylabel = "Predicción precio")
```

### Analisis Algortimo Regresión Lineal: Complejidad Algorítmica

En un algoritmo de regresión de mínimos cuadrados, que tengo que hacer las siguientes operaciones para calcular los coeficientes de regresión:

Para un mínimo de cuadrados de regresión con **n** ejemplos de formación y **c** características, se necesita:

- La multiplicación de la matriz, la complejidad: $\mathcal{O}(c^{2}n)$
- La inversión de matrices, la complejidad: $\mathcal{O}(c^{3})$
- La multiplicación de la matriz, la complejidad: $\mathcal{O}(c^{2}n)$
- La multiplicación de la matriz, la complejidad: $\mathcal{O}(cn)$

Teniendo en cuento todos los diferentes aspectos de complejidad por cada uno de los procesos que se realizan la complejidad del algoritmo de la regresión lineal por mínimos cuadrados sería $$\mathcal{O}(c^{2}n)$$
